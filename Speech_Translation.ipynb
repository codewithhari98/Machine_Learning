```python
#import packages
from gtts import gTTS, lang
import os
from tkinter import *
from tkinter import messagebox
import speech_recognition as sr
import text2text as t2t
t2t.Transformer.PRETRAINED_TRANSLATOR = "facebook/m2m100_418M" #Remove this line for the larger model
#import PyAudio

#define functions
#text to speech conversion
def text_to_speech(text_sample='No text given',axnt='en'):
        #read inputs given by user
        text = str(text_sample)
        
        language = str(axnt[0:2])

        print(language)
        
        if (len(text)<=1) | (len(language)<=0):
                messagebox.showerror(message="Enter required details")
                return
             
        #Using the inputs, convert the text to speech
        speech = gTTS(text = text, lang = language, slow = False)
        #save the speech to an MP3 file
        speech.save("text.mp3")
        #Play the file usinf mpg123 in linux and start in windows
        os.system("mpg123 "+"text.mp3")

#List the supported languages and their keys
def list_languages():
        #access languages and access codes using lang.tts_langs()
        messagebox.showinfo(message=list(lang.tts_langs().items()))

#speech to text conversion
def speech_to_text(lang1):
        recorder = sr.Recognizer()

        messagebox.showinfo(message="Speak into the microphone and wait after finishing the recording")   
        with sr.Microphone() as mic:  
                #Prompt the user to record
                #Record audio from the user
                recorder.adjust_for_ambient_noise(mic)
                audio_input = recorder.listen(mic)    
                try:                        #Convert to text
                        text_output =recorder.recognize_google(audio_input,language=str(lang1))
                        #Display the output
                        messagebox.showinfo(message="You said:\n "+text_output)        
                except:
                         messagebox.showerror(message="Couldn't process the audio input.")
        
        return text_output

def translation(txt,lang1,lang2):
        ### Text Handler API quick start
        h = t2t.Handler([txt], src_lang=str(lang1)) #Initialize with some text
        h.tokenize() #[['▁Hello', ',', '▁World', '!']]
        trans = h.translate(tgt_lang=str(lang2))
        return trans


def speech_to_speech(lang1,lang2):
        lang1 = lang1[0:2]
        lang2 = lang2[0:2]

        txt = speech_to_text(lang1)
        print('txt before translation',txt)
        txt_after = translation(txt,lang1,lang2)
        print('text after translation',txt_after)
        text_to_speech(txt_after,lang2)

def tts_box():
        #Invoke call to class to view a window
        window = Tk()
        #Set dimensions of window and title
        window.geometry("500x300")
        window.title("TEXT TO SPEECH")
        title_label = Label(window, text="TEXT TO SPEECH").pack()

        button1 = Button(window,text='List languages', bg = 'Turquoise',fg='Red',command=list_languages).place(x=10,y=210)

        #text_to_speech input
        text_label = Label(window, text="Text:").place(x=10,y=20)
        text_entry = Text(window, width=30,height=5)
        text_entry.place(x=80,y=20)
        #Accent input
        accent_label = Label(window, text="Accent:").place(x=10,y=110)
        accent_entry = Text(window, width=5,height=1)
        accent_entry.place(x=80,y=110)

        button6 = Button(window,text='Speak', bg = 'Turquoise',fg='Red',command=lambda:text_to_speech(text_entry.get("1.0","end"),accent_entry.get("1.0","end"))).place(x=150,y=250)


def translate_box():
        #Invoke call to class to view a window
        window = Tk()
        #Set dimensions of window and title
        window.geometry("500x300")
        window.title("SPEECH TRANSLATION")
        title_label = Label(window, text="SPEECH TRANSLATION").pack()

        button1 = Button(window,text='List languages', bg = 'Turquoise',fg='Red',command=list_languages).place(x=10,y=120)
        #Language1
        lang1 = Label(window, text="Translate From:").place(x=10,y=20)
        lang1_entry = Text(window, width=5,height=1)
        lang1_entry.place(x=120,y=20)

        #Language2
        lang2 = Label(window, text="Translate To:").place(x=10,y=70)
        lang2_entry = Text(window, width=5,height=1)
        lang2_entry.place(x=120,y=70)

        button5 = Button(window,text='Translate', bg = 'Turquoise',fg='Red',command=lambda:speech_to_speech(lang1_entry.get("1.0","end"),lang2_entry.get("1.0","end"))).place(x=150,y=150)
        #speech_to_speech()

        # #close the app
        # window.mainloop()

def main():

        #Invoke call to class to view a window
        window = Tk()
        #Set dimensions of window and title
        window.geometry("500x300")
        window.title("DASHBOARD")

        #Perform the functions
        #button1 = Button(window,text='List languages', bg = 'Turquoise',fg='Red',command=list_languages).place(x=200,y=50)
        button2 = Button(window,text='Text to Speech', bg = 'Turquoise',fg='Red',command=tts_box).place(x=200,y=80)
        button3 = Button(window,text='Speech to Text', bg = 'Turquoise',fg='Red',command=lambda:speech_to_text('en')).place(x=200,y=140)
        button4 = Button(window,text='Translate Languages', bg = 'Turquoise',fg='Red',command=translate_box).place(x=200,y=200)

        #close the app
        window.mainloop()


if __name__ == "__main__":
        main()


```
